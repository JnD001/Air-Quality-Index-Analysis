```python
columns=['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene']

for col in columns:
    data_cleaned[col] = data_cleaned[col].fillna(data_cleaned[col].median())

print(data_cleaned[columns].isnull().sum())
```

```python
buckets = ['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe']

max_min_value={}

for bucket in buckets:
    
    bucket_data=data_cleaned[data_cleaned['AQI_Bucket']==bucket]

    min_aqi=bucket_data['AQI'].min()
    max_aqi=bucket_data['AQI'].max()

    max_min_value[bucket]={"Min AQI" : min_aqi,"Max AQI": max_aqi}
    
for bucket, values in max_min_value.items():
    print(f"{bucket} : \n  Min AQI : {values['Min AQI']}, Max AQI : {values['Max AQI']}")
```

```python
data_cleaned['AQI'] = data_cleaned['AQI'].fillna(data_cleaned['AQI'].median())
data_cleaned.head()
```

```python
data_cleaned['AQI_Bucket'] = data_cleaned.apply(lambda row: 'Moderate' if pd.isna(row['AQI_Bucket']) and row['AQI'] == 116.0 else row['AQI_Bucket'], axis=1)

data_cleaned.isnull().sum()
```

```python
# Select numeric columns
numeric_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene']

# Iteratively remove outliers
while True:
    outlier_mask = pd.Series(False, index=data_cleaned.index)  # Reset mask
    for col in numeric_columns:
        Q1, Q3 = data_cleaned[col].quantile([0.25, 0.75])  # Calculate Q1 and Q3
        IQR = Q3 - Q1
        lower_bound, upper_bound = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
        outlier_mask |= (data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound)
    
    total_outliers = outlier_mask.sum()
    if total_outliers == 0:  # Stop when no more outliers
        break
    data_cleaned = data_cleaned[~outlier_mask]

# Report final outliers
final_outliers = 0
for col in numeric_columns:
    Q1, Q3 = data_cleaned[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_bound, upper_bound = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    num_outliers = ((data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound)).sum()
    print(f"Column: {col}")
    print(f"Number of Outliers: {num_outliers}")
    final_outliers += num_outliers

# Final summary
if final_outliers == 0:
    print("No outliers detected in the cleaned dataset.")
else:
    print(f"Total Outliers Remaining: {final_outliers}")
```

```python
data_cleaned.columns = data_cleaned.columns.str.lower().str.replace(' ', '_')
data_cleaned['datetime'] = pd.to_datetime(data_cleaned['datetime'])
data_cleaned.info()
```

```python
print(data_cleaned['city'].unique())
print(data_cleaned['aqi_bucket'].value_counts())
```

```python
# Selecting numeric columns
numeric_columns = data_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Setting up the grid size
n_cols = 3  # Number of columns in the grid
n_rows = -(-len(numeric_columns) // n_cols)  # Calculate rows based on the number of columns

# Creating subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
axes = axes.flatten()  # Flatten axes for easier iteration

for i, column in enumerate(numeric_columns):
    data_cleaned[column].hist(bins=20, ax=axes[i], color='skyblue', edgecolor='black')
    axes[i].set_title(f'{column} Distribution', fontsize=10)
    axes[i].set_xlabel(column, fontsize=8)
    axes[i].set_ylabel('Frequency', fontsize=8)

# Turn off unused axes
for i in range(len(numeric_columns), len(axes)):
    axes[i].axis('off')

plt.tight_layout()  # Adjust layout to avoid overlapping
plt.show()
```

```python
data_cleaned.groupby('datetime')['aqi'].mean().plot()
plt.title('AQI Over Time', fontsize=14)
plt.xlabel('Datetime', fontsize=12)
plt.ylabel('Average AQI', fontsize=12)
plt.show()
```

```python
# Optional: Sampling for large datasets
sampled_data = data_cleaned[['pm2.5', 'pm10', 'no', 'no2', 'aqi']].sample(1000, random_state=42)

# Pair plot with customization
sns.pairplot(
    sampled_data,
    diag_kind='kde',  # Smoother diagonal plots
    kind='scatter',   # Default scatter plot for pairwise relationships
    plot_kws={'alpha': 0.6, 's': 20, 'edgecolor': 'w'},  # Scatter plot transparency and size
    diag_kws={'fill': True},  # KDE customization
)

plt.suptitle('Pairwise Relationships of Selected Pollutants and AQI', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Group by city and aqi_bucket, then calculate the mean AQI
aqi_by_city = data_cleaned.groupby(['city', 'aqi_bucket'])['aqi'].mean().unstack()

# Plotting the stacked bar chart
ax = aqi_by_city.plot(kind='bar', stacked=True, figsize=(10, 6))

# Title and labels
plt.title('AQI by City and Bucket', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Average AQI', fontsize=12)


# Show the plot
plt.show()
```
