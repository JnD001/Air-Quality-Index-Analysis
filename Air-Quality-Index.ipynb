```python
data_cleaned=data_cleaned.loc[data_cleaned.isnull().mean(axis=1)*100 <= 50]
```

```python
data_cleaned.head()
```

```python
columns=['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene']

for col in columns:
    data_cleaned[col] = data_cleaned[col].fillna(data_cleaned[col].median())

print(data_cleaned[columns].isnull().sum())
```

```python
buckets = ['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe']

max_min_value={}

for bucket in buckets:
    
    bucket_data=data_cleaned[data_cleaned['AQI_Bucket']==bucket]

    min_aqi=bucket_data['AQI'].min()
    max_aqi=bucket_data['AQI'].max()

    max_min_value[bucket]={"Min AQI" : min_aqi,"Max AQI": max_aqi}
    
for bucket, values in max_min_value.items():
    print(f"{bucket} : \n  Min AQI : {values['Min AQI']}, Max AQI : {values['Max AQI']}")
```
